```python
x = df['Total Normal Rainfall (June\'2018 to May\'2019) in mm']
y = df['Total Actual Rainfall (June\'2018 to May\'2019) in mm']
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.75, shuffle = True)
from sklearn.linear_model import LinearRegression
print(x_train.shape)
print(x_test.shape)
```

#Model Building using Linear Regression

```python
import numpy as np
x_train=np.array(x_train).reshape(-1,1)
y_train=np.array(y_train).reshape(-1,1)

model=LinearRegression()
model.fit(x_train,y_train)

print('The Coeffecient is',model. coef_)
print('The interecept is',model.intercept_)
```

```python
y_test=np.array(y_test).reshape(-1,1)
y_pred=model.predict(y_test)
print(y_pred)
```

```python
plt.scatter(x_test, y_test, color = 'y', marker = '.')
plt.plot(x_test, y_pred, color='r')
```

#Performance Metrics

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import math

print('Mean absolute error:', mean_absolute_error(y_test, y_pred))

print("Mean squared error:", mean_squared_error(y_test, y_pred))
print('variance score:',r2_score(y_test, y_pred))

print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, y_pred)))
```
